# Join operators

```python
# Left + common from right

left= pd.DataFrame({'A':['A0','A1','A3'],'B':['B0','B1','B3']  }, index=['K0','K1','K3'])
left
```

```python
right = pd.DataFrame({'C':['C0','C2','C3'], 'D':['D0','D2','D3']   }, index= ['K0','K2','K3' ])
right
```

```python
left.join(right)
```

```python
left.join(right , how='inner')
```

```python
left.join(right , how='outer')
```

```python
# use of apply function

import pandas as pd
```

```python
# create a function and apply it

def salary_hike(x):
    if x>3000:
        return np.ceil(x + (x * .3))
    else: return( x * 1.5)
```

```python
df.head(10)
```

```python
df['New_Salary']= df['Salary'].apply(salary_hike)
```

```python
df
```

```python
df['Name_length']= df['Emp_Name'].apply(len)
```

```python
# sq of length
df['len_sq']= df['Name_length'].apply(lambda x : np.square(x))
```
