```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option('display.max_columns',None)
```

```python
df= pd.read_csv('titanic.csv')
```

```python
df.head(5)
```

```python
sns.lineplot(x='Age',y='Pclass',hue='Sex',style='Sex', markers=True,data=df)
```

# Box Plot

```python
sns.barplot(x='Survived',y='Age', hue="Sex",data=df, color='orange')
plt.show()
```

# ScatterPlot

```python
sns.scatterplot(x="Age",y="Fare",hue="Fare",data=df, style='Sex')
```

# Histogram / Distplot

```python
sns.distplot(df['Age'], hist=True,kde=False, bins=50,color='orange' )
```

```python
sns.distplot(df['Age'], hist=False,kde=True, bins=50,color='orange' )
```

```python
sns.distplot(df['Age'], hist=True,kde=True, bins=50,color='orange', vertical=True )
```

# JointPlot

```python
sns.jointplot(x='Age',y='Fare',data=df)
```

```python
sns.jointplot(x='Age',y='Fare',data=df, color='olive')
```

```python
sns.jointplot(x='Age',y='Fare',data=df, color='olive', kind='reg')
```

# Boxplot

```python
sns.boxplot(y='Age',x='Survived', data=df)
```

```python
sns.boxplot(y='Age',x='Survived', data=df, palette='Set1')
```

```python
sns.boxplot(y='Age',x='Survived', data=df, palette='Set1', linewidth=3)
```

```python
sns.boxplot(y='Age',x='Survived', data=df, palette='Set1', linewidth=3, hue='Sex')
```
